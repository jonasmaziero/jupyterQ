### Estatística de notas versus número de faltas <br> Disciplina de Física Matemática II, UFSM

```python
%run jupyterQ.ipynb
from scipy import optimize
```

```python
def test_func(x, a, b):
    return a + b*x
```

```python
# 2014/2
faltas = np.array([8,4,6,12,23,4,11,2,2,14,12,12,6])
notas = np.array([7.77,5.1,7.52,5.52,0.96,9.16,3.82,7.99,8.46,5,8.52,4.17,5.02])
params, params_covariance = optimize.curve_fit(test_func, faltas, notas, p0=[2, 2])
print(params)
plt.plot(faltas, test_func(faltas, params[0], params[1]), color = 'cyan',label='ajuste')
plt.scatter(faltas, notas, color = 'blue', label='dados')
plt.legend(loc='best')
plt.xlabel('Faltas')
plt.ylabel('Notas')
axes = plt.gca()
axes.set_xlim([0,60])
axes.set_ylim([-1,11])
plt.show()
```
