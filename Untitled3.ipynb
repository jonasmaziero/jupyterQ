```python
#### Purificação do estado tipo werner para 1 qubit

#aqui calculei os angulos diretamente
from qiskit import * #from qiskit import everything
%matplotlib inline
from qiskit.quantum_info import Statevector
from qiskit.quantum_info import partial_trace
from qiskit.tools.visualization import plot_histogram 
from qiskit.quantum_info import purity

import numpy as np


circuit = QuantumCircuit(2,2) # cria um circuito quântico com 2qubits e 2cbits
for y in np.arange(np.pi/2, np.pi, np.pi/8):
    for x in np.arange(0, np.pi/2, np.pi/8):
        # circuito para obter o estado purificado
        circuit.u3(x,0,0,0)
        circuit.u3(y,0,0,1)
        circuit.cx(1,0)
        circuit.cz(1,0)
        sv = Statevector.from_label('00') #cria o estado |0,0>
        new_sv = sv.evolve(circuit) # evolui o estado |0,0> de acordo com o circuito criado
        rhoA = partial_trace(new_sv,[1])
        circuit.draw()
        print(rhoA)
        print(purity(rhoA))
```

```python
#### Purificação do estado tipo werner para 1 qubit

#aqui calculei os angulos a partir de x,w
from qiskit import * #from qiskit import everything
%matplotlib inline
from qiskit.quantum_info import Statevector
from qiskit.quantum_info import partial_trace
from qiskit.tools.visualization import plot_histogram 
from qiskit.quantum_info import purity

import numpy as np


circuit = QuantumCircuit(2,2) # cria um circuito quântico com 2qubits e 2cbits
for y in np.arange(0, 1, 0.2):
    for x in np.arange(0, 1, 0.2):
        # circuito para obter o estado purificado
        circuit.u3(2*np.arcsin(np.sqrt(x)),0,0,0)
        circuit.u3(np.arccos(-y),0,0,1)
        circuit.cx(1,0)
        circuit.cz(1,0)
        sv = Statevector.from_label('00') #cria o estado |0,0>
        new_sv = sv.evolve(circuit) # evolui o estado |0,0> de acordo com o circuito criado
        rhoA = partial_trace(new_sv,[1])
        print(rhoA)
        print(purity(rhoA))
```
